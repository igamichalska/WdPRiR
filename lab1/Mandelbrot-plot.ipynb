```python
import matplotlib.pyplot as plt
```

```python
pixels = [32, 64, 128, 256, 512, 1024, 2048, 4096, 8192]
```

```python
file = open("output.txt", "r")
times = []
for line in file:
        times.append(float(line.strip('\n')))
```

```python
plt.scatter(pixels,times)
plt.xlabel('Liczba pikseli')
plt.ylabel('Średni czas [s]')
plt.title('Średni czas generacji zbioru Mandelbrota')
plt.savefig('plot.png')
plt.show()
```
