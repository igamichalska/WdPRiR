```python
import matplotlib.pyplot as plt
```

```python
pixels = [16, 32, 64, 128, 256, 512, 1024, 2048, 4096, 8192]
```

```python
file = open("output-threads.txt", "r")
threads = []
for line in file:
        threads.append(float(line.strip('\n')))
```

```python
plt.scatter(pixels, threads)
plt.loglog()
plt.xlabel('Liczba pikseli')
plt.ylabel('Średni czas [s]')
plt.title('Średni czas generacji zbioru Mandelbrota - wątki')
plt.show()
```

```python
file = open("output-pools-new01-01.txt", "r")
pools01 = []
for line in file:
        pools01.append(float(line.strip('\n')))
```

```python
plt.scatter(pixels, pools01)
plt.loglog()
plt.xlabel('Liczba pikseli')
plt.ylabel('Średni czas [s]')
plt.title('Pula wątków - tworzenie tylko raz')
plt.show()
```

```python
file = open("output-pools-new02-02.txt", "r")
pools02 = []
for line in file:
        pools02.append(float(line.strip('\n')))
```

```python
plt.scatter(pixels,pools02)
plt.loglog()
plt.xlabel('Liczba pikseli')
plt.ylabel('Średni czas [s]')
plt.title('Pula wątków - tworzenie za każdym razem')
plt.show()
```

```python
plt.scatter(pixels, threads, label = 'threads')
plt.scatter(pixels, pools01, label = 'pools 1')
plt.scatter(pixels, pools02, label = 'pools 2')
plt.loglog()
plt.legend()
plt.show()
```
